```python
for key in file_dl_dict:
    dl_name = file_dl_dict[key][0][0] + ".xlsx"
    dl_url = "https://www.rba.gov.au/statistics/tables/xls/" + dl_name
    r = requests.get(dl_url)
    with open("./Data/" + dl_name, 'wb') as f:
              f.write(r.content)    
            
            # can i put extra info into the dictionary? i.e. the required columns + rows to keep
```

```python
# separate functions: download the files, extract all the series + metadata, clean/format the data, visualise the data

output_d = {} # For storing all outputs in

for key in file_dl_dict:   
    data_name = str(file_dl_dict[key][0][0])
    row_filter = file_dl_dict[key][1]
    series_filter = file_dl_dict[key][2]
    
    read_path = './data/' + data_name + '.xlsx'
    rba_data = pd.read_excel(read_path, sheet_name = 'Data', skiprows = 1)

    # Extract Publication Date
    print('Publication date: ', rba_data.loc[rba_data['Title'] == 'Publication date'].values[0,1]) #loc = index check
    
    # should put publication dates into another list
    
    # Transform Data 

    # Drop Rows
    rba_data_fmt = rba_data[~rba_data['Title'].isin(row_filter)]
    rba_data_fmt = rba_data_fmt.dropna(subset=['Title']) # Drop these rows 
    
    # rba_data_fmt.head()
    series_id = rba_data[rba_data['Title'] == 'Series ID'].values[0] # Series ID values
    units = rba_data[rba_data['Title'] == 'Units'].values[0] # Unit values
    title = rba_data_fmt.columns.values # Series ID Names
    series_lookup = pd.DataFrame(list(zip(series_id,title,units)), columns=['Series ID','Title','Unit'])
    series_lookup = series_lookup[series_lookup['Series ID'] != 'Series ID'] # Need to save these in the dictionary output

    # select the series_to_filter
    series_filter = series_filter # Series ID to lookup
    column_filter = series_lookup.loc[series_lookup['Series ID'].isin(series_filter), 'Title'].values.tolist()
    column_filter.insert(0, 'Title')

    # Filter Columns based on column_filter
    rba_data_fmt = rba_data_fmt.loc[rba_data_fmt['Title']!='Series ID',column_filter]
    rba_data_fmt = rba_data_fmt.reset_index(drop=True)
    rba_data_fmt = rba_data_fmt.rename(columns={'Title': 'Date'})
    rba_data_fmt = rba_data_fmt.convert_dtypes() # Convert variable types

    output_d[data_name] = rba_data_fmt # Need to save this
```

```python
# Visualise the trend

rba_c_1_2 = output_d['c01-2-hist']
rba_c_1_2 = rba_c_1_2[rba_c_1_2['Date'] >= '2018-01-01'] # In order to calculate MoM + YoY
rba_c_1_2['Date_Axis'] = rba_c_1_2['Date'].dt.strftime("%b-%y")
rba_c_1_2['Year'] = rba_c_1_2['Date'].dt.year
rba_c_1_2['Month'] = rba_c_1_2['Date'].dt.strftime("%b")
rba_c_1_2.rename(columns = {'Number of accounts: personal cards' : 'Accounts',
                            'Total balances: personal cards' : 'Balance', 
                            'Balances accruing interest: personal cards' : 'Interest Balance',
                            'Credit limits: personal cards' : 'Limit',
                            'Total value of transactions: personal cards' : 'Transaction Value'}, inplace = True)
rba_c_1_2['Utilisation'] = (rba_c_1_2['Balance'] / rba_c_1_2['Limit'])
rba_c_1_2['Balance MoM'] = rba_c_1_2['Balance'] - rba_c_1_2['Balance'].shift(periods=1)
rba_c_1_2['Balance MoM Percent'] = rba_c_1_2['Balance']/rba_c_1_2['Balance'].shift(periods=1)-1
rba_c_1_2['Balance YoY'] = rba_c_1_2['Balance'] - rba_c_1_2['Balance'].shift(periods=12)
rba_c_1_2['Balance YoY Percent'] = rba_c_1_2['Balance']/rba_c_1_2['Balance'].shift(periods=12)-1
rba_c_1_2['Interest Balance Percent'] = rba_c_1_2['Interest Balance']/rba_c_1_2['Balance']
rba_c_1_2['Transaction Value YoY'] = rba_c_1_2['Transaction Value']/rba_c_1_2['Transaction Value'].shift(periods=12)-1
rba_c_1_2 = rba_c_1_2[rba_c_1_2['Date'] >= '2019-01-01'] 
```

```python
# Summary stats 

# Pull out both latest figures + data needed for commentary

max_date = max(rba_c_1_2['Date'])
latest_balance = rba_c_1_2[rba_c_1_2['Date'] == max_date]['Balance'].values[0]
latest_balance_fmt = "$" + str(round(latest_balance,-2)/1000) + "B"

# Balances
print("In " + max_date.strftime("%b-%y") + " Credit Card Balances were " + latest_balance_fmt)

```

```python
f, axes = plt.subplots(2,2, figsize=(15,12))

max_y_balance = int(round(max(rba_c_1_2['Balance']),-4))
max_y_trans = int(round(max(rba_c_1_2['Transaction Value']),-3))

# Credit Card Balance
balance_graph = sns.lineplot(x='Date_Axis', y='Balance', data=rba_c_1_2, color = 'red', ax=axes[0,0], label='Balance (LHS)')
balance_graph.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '${:,.0f}'.format(y/1000) + 'B')) # Not sure how to set this programmatically

# Credit Card Utilisation Rate (Balance/Limit)
balance_graph2 = balance_graph.twinx()
balance_graph2.plot(balance_graph.get_xticks(), rba_c_1_2['Utilisation'], label='Utilisation (RHS)')
balance_graph2.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y*100) + '%'))

# Credit Card Balance YoY Change
palette_custom = ['#B0B8B4FF', '#FC766AFF', '#184A45FF']
balance_YoY_graph = sns.barplot(x='Date_Axis', y='Balance YoY Percent', data=rba_c_1_2, hue='Year', palette = palette_custom, 
                                dodge=False, ax=axes[0,1])
balance_YoY_graph.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y*100) + '%'))

# Credit Card Balance Accruing Interest
balance_interest_graph = sns.lineplot(x='Date_Axis', y='Interest Balance', data=rba_c_1_2, color ='red', ax=axes[1,0], 
                                      label='Interest Balance (LHS)')
balance_interest_graph.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '${:,.0f}'.format(y/1000) + 'B'))

# % of Credit Card Balance Accruing Interest
balance_interest_graph2 = balance_interest_graph.twinx()
balance_interest_graph2.plot(balance_interest_graph2.get_xticks(), rba_c_1_2['Interest Balance Percent'], 
                             label='% of Total Balance')
balance_interest_graph2.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '{:,.0f}'.format(y*100) + '%'))

# Transaction Volumes
trans_val_graph = sns.lineplot(x='Month', y='Transaction Value', data=rba_c_1_2, hue = 'Year', palette = palette_custom,
                               ax=axes[1,1])
trans_val_graph.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: '${:,.0f}'.format(y/1000) + 'B'))

graphs = []

# Use value NONE where invalid
graphs.append({
    'graph': balance_graph,
    'title': 'Credit Card Balance',
    'label-x': 'Date',
    'label-y': 'Balance',
    'legend': 'upper right',
    'limit-x-lower': 'NONE',
    'limit-x-upper': 'NONE',  
    'limit-y-lower': 0,
    'limit-y-upper': max_y_balance,  
    'rotate-x-ticks': 45,
    'date-axis-spread': 'Y',
    'secondary-graph': balance_graph2,
    'secondary-label-y': 'Utilisation',
    'secondary-legend': 'lower right',
    'secondary-limit-y-lower': 0,
    'secondary-limit-y-upper': 1,
})

graphs.append({
    'graph': balance_YoY_graph,
    'title': 'Credit Card Balance - YoY Changes',
    'label-x': 'Date',
    'label-y': 'YoY Balance Change',
    'legend': 'NONE',
    'limit-x-lower': 'NONE',
    'limit-x-upper': 'NONE',  
    'limit-y-lower': 'NONE',
    'limit-y-upper': 'NONE',  
    'rotate-x-ticks': 45,
    'date-axis-spread': 'Y',
    'secondary-graph': 'NONE',
    'secondary-label-y': 'NONE',
    'secondary-legend': 'NONE',
    'secondary-limit-y-lower': 'NONE',
    'secondary-limit-y-upper': 'NONE',
})

graphs.append({
    'graph': balance_interest_graph,
    'title': 'Credit Card Balance - Balances Accruing Interest',
    'label-x': 'Date',
    'label-y': 'Interest Balance',
    'legend': 'upper right',
    'limit-x-lower': 'NONE',
    'limit-x-upper': 'NONE',  
    'limit-y-lower': 0,
    'limit-y-upper': max_y_balance,  
    'rotate-x-ticks': 45,
    'date-axis-spread': 'Y',
    'secondary-graph': balance_interest_graph2,
    'secondary-label-y': '% of Total Balance',
    'secondary-legend': 'lower right',
    'secondary-limit-y-lower': 0,
    'secondary-limit-y-upper': 1,
})

graphs.append({
    'graph': trans_val_graph,
    'title': 'Credit Card - Transaction Values',
    'label-x': 'Month',
    'label-y': 'Transaction Balance',
    'legend': 'NONE',
    'limit-x-lower': 'NONE',
    'limit-x-upper': 'NONE',  
    'limit-y-lower': 0,
    'limit-y-upper': max_y_trans,  
    'rotate-x-ticks': 45,
    'date-axis-spread': 'Y',
    'secondary-graph': 'NONE',
    'secondary-label-y': 'NONE',
    'secondary-legend': 'NONE',
    'secondary-limit-y-lower': 'NONE',
    'secondary-limit-y-upper': 'NONE',
})

# need to add a counter here
for graph in graphs: #Loop to set all graph parameters
    graph['graph'].set_title(graph['title'], loc='left') 
    graph['graph'].set(xlabel=graph['label-x'], ylabel=graph['label-y'])
    
    if graph['legend'] != 'NONE': # Legend
        graph['graph'].legend(loc=graph['legend']) 
    if graph['limit-x-lower'] != 'NONE': # X Limits
        graph['graph'].set_xlim(graph['limit-x-lower'], graph['limit-x-upper']) 
    if graph['limit-y-lower'] != 'NONE': # Y Limits
        graph['graph'].set_ylim(graph['limit-y-lower'], graph['limit-y-upper'])       
    if graph['rotate-x-ticks'] != 'NONE': # X Tick Rotation
        graph['graph'].xaxis.set_tick_params(labelrotation=graph['rotate-x-ticks'])
    if graph['date-axis-spread'] == 'Y':
        for ind, label in enumerate(graph['graph'].get_xticklabels()):
            if ind % 3 == 0:  # every 3rd label is kept
                label.set_visible(True)
            else:
                label.set_visible(False)
    
    if graph['secondary-graph'] != 'NONE':
        graph['secondary-graph'].set(ylabel=graph['secondary-label-y'])
        
        if graph['secondary-legend'] != 'NONE':
            graph['secondary-graph'].legend(loc=graph['secondary-legend']) 
        if graph['secondary-limit-y-lower'] != 'NONE':
            graph['secondary-graph'].set_ylim(graph['secondary-limit-y-lower'], graph['secondary-limit-y-upper'])     

plt.tight_layout()
plt.show()
```
